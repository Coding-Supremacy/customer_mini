```python
# 가입일과 고객 세그먼트의 관계 확인
import matplotlib.pyplot as plt
import seaborn as sb

#한글 폰트 설정
plt.rc('font', family='NanumGothic')


# 고객 세그먼트별 가입일 분포 시각화
plt.figure(figsize=(10, 6))
sb.countplot(x='가입연도', hue='고객 세그먼트 (Customer Segment)', data=df)
plt.title('가입연도와 고객 세그먼트 관계')
plt.xlabel('가입연도')
plt.ylabel('가입 수')
plt.legend(title='고객 세그먼트')
plt.savefig('가입연도와 고객 세그먼트 관계.png')
plt.show()


# 가입월과 고객 세그먼트의 관계 확인
plt.figure(figsize=(10, 6))
sb.countplot(x='가입월', hue='고객 세그먼트 (Customer Segment)', data=df)
plt.title('가입월과 고객 세그먼트 관계')
plt.xlabel('가입월')
plt.ylabel('가입 수')
plt.legend(title='고객 세그먼트')
plt.savefig('가입월와 고객 세그먼트 관계.png')
plt.show()
```

```python
#구매한 제품과 차량 구분은 불일치함.
#G80 (RG3) 는 준중형 세단
#모델명은 유의미한 데이터가 아니므로 차량구분(vehicle types)으로 분석 진행
#대신 모델 출시일을 기준으로 자동차가 언제 출시되었는지를 확인
```

```python
# 모델과 출시 년월 데이터
launch_dates = {
    'G70 (IK)': '2017-09',
    'Santa-Fe ™': '2018-01',  # 첫 출시
    'NEXO (FE)': '2018-01',
    'Avante (CN7 N)': '2020-05',
    'G80 (RG3)': '2020-03',
    'Grandeur (GN7 HEV)': '2022-01',
    'IONIQ (AE EV)': '2016-01',
    'i30 (PD)': '2017-03',
    'Palisade (LX2)': '2018-12',
    'Tucson (NX4 PHEV)': '2021-05',
    'Avante (CN7 HEV)': '2020-07',
    'IONIQ 6 (CE)': '2022-06',
    'G90 (HI)': '2022-03',
    'Santa-Fe (MX5 PHEV)': '2022-06',
    'G90 (RS4)': '2022-03'
}

# '구매한 제품 (Purchased Product)'에 따라 출시 년월 컬럼 추가
df['제품 출시년월 (Launch Date)'] = df['구매한 제품 (Purchased Product)'].map(launch_dates)

# '제품 출시년월 (Launch Date)'를 년, 월로 분리
df['출시연도'] = pd.to_datetime(df['제품 출시년월 (Launch Date)'], errors='coerce').dt.year
df['출시월'] = pd.to_datetime(df['제품 출시년월 (Launch Date)'], errors='coerce').dt.month
```

```python
df
```

```python
# 생년월일을 datetime 형식으로 변환
df['생년월일 (Date of Birth)'] = pd.to_datetime(df['생년월일 (Date of Birth)'], errors='coerce')

# 2025년을 기준으로 나이 계산
df['나이'] = 2025 - df['생년월일 (Date of Birth)'].dt.year

# 나이를 기준으로 연령대 분류 (10대, 20대, 30대 ... 80대)
df['연령대'] = (df['나이'] // 10) * 10

# 결과 확인
df[['생년월일 (Date of Birth)', '나이', '연령대']].head()
```

```python
filtered_customers=df.loc[: ,['성별 (Gender)','차량구분(vehicle types)','거래 금액 (Transaction Amount)','제품 구매 빈도 (Purchase Frequency)',]]
```

```python
# 연령대 컬럼 이동
cols = filtered_customers.columns.tolist()
cols.remove('연령대')
cols.insert(1, '연령대')
filtered_customers = filtered_customers[cols]
filtered_customers
```

```python
#생년 삭제
filtered_customers.drop('생년 (Year of Birth)', axis=1, inplace=True)
filtered_customers
```

```python
# 제품 구매 날짜 (Purchase Date) 년, 월 별도 컬럼으로 분리
filtered_customers['구매 년도'] = filtered_customers['제품 구매 날짜 (Purchase Date)'].dt.year
filtered_customers['구매 월'] = filtered_customers['제품 구매 날짜 (Purchase Date)'].dt.month
#제품 구매 날짜 (Purchase Date) 컬럼 삭제
filtered_customers.drop('제품 구매 날짜 (Purchase Date)', axis=1, inplace=True)
filtered_customers
```

```python
#제품 출시년월 년, 월로 분리
filtered_customers['출시 년도'] = filtered_customers['제품 출시년월 (Launch Date)'].str.split('-', expand=True)[0]
filtered_customers['출시 월'] = filtered_customers['제품 출시년월 (Launch Date)'].str.split('-', expand=True)[1]
#제품 출시년월 (Launch Date) 컬럼 삭제
filtered_customers.drop('제품 출시년월 (Launch Date)', axis=1, inplace=True)
filtered_customers
```

```python
#filtered_customers 저장
filtered_customers.to_csv('filtered_customers.csv', index=False)
```
