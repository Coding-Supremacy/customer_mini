```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('고객데이터_클러스터링.csv')
```

```python
df.head()
```

```python
#신규고객이 어떤 클러스터에 속하나 확인하는 클래시파이어 모델만들기
# 1. 데이터 전처리
# 2. 모델링
# 3. 모델평가
# 4. 예측
# 5. 신규고객 예측
# 6. 모델저장 
```

```python
# 1. 데이터 전처리
num_features = ['거래 금액 (Transaction Amount)', '제품 구매 빈도 (Purchase Frequency)', '연령대']
str_features = ['성별 (Gender)', '차량구분(vehicle types)', '거래 방식 (Transaction Method)', '제품 출시년월 (Launch Date)']
```

```python
df.head()
```

```python
#스케일링, 인코딩 전처리
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('str', OneHotEncoder(handle_unknown='ignore'), str_features)
    ])
```

```python
from sklearn.linear_model import LogisticRegression
#선형방정식 기준 모델 : 로지스틱 회귀
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42))
])
```

```python
from sklearn.ensemble import RandomForestClassifier
# RandomForestClassifier 여러개의 의사결정나무를 만들어서 투표를 통해 결과를 결정하는 방식
from sklearn.pipeline import Pipeline
pipeline2 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier2', RandomForestClassifier(random_state=42))
])
```

```python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
#SVC : 서포트 벡터 머신, 데이터를 분류하는 경계선을 정의하는 방식
pipeline3 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier3', SVC(random_state=42))
])
```

```python
#X, y 분리
X = df[num_features + str_features]
y = df['cluster']
```

```python
#학습용 데이터와 테스트용 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)

#모델 학습
pipeline.fit(X_train, y_train)
pipeline2.fit(X_train, y_train)
pipeline3.fit(X_train, y_train)
```

```python
#모델 평가
from sklearn.metrics import accuracy_score
y_pred = pipeline.predict(X_test)
y_pred2 = pipeline2.predict(X_test)
y_pred3 = pipeline3.predict(X_test)
```

```python
#성능 평가
print('로지스틱 회귀 정확도: ', accuracy_score(y_test, y_pred))
print('랜덤 포레스트 정확도: ', accuracy_score(y_test, y_pred2))
print('SVM 정확도: ', accuracy_score(y_test, y_pred3))
```

```python
#에러 확인
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred2))
```

```python
print(classification_report(y_test, y_pred3))
```
